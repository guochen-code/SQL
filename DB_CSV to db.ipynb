# import csv to db

```python
# create db
from pathlib import Path
Path('my_data.db').touch()
```

```python
# connect with empty db
import sqlite3
conn = sqlite3.connect('my_data.db')
c = conn.cursor()
```

```python
# create table
c.execute('''CREATE TABLE users (name text, population int)''')
```

```python
import pandas as pd
# load the data into a Pandas DataFrame
# Pandas makes it easy to load this CSV data into a sqlite table:
users = pd.read_csv('city.csv')
# write the data to a sqlite table
# The to_sql method makes it easy to write DataFrames to databases.
users.to_sql('users', conn, if_exists='append', index = False)
```

```python
# The fetchall() method returns an array of tuples. Here too long, clear output
c.execute('''SELECT * FROM users''').fetchall() # [('Abilene', 115930), ('Akron', 217074)]
```

```python
pd.read_sql('''SELECT * FROM users''', conn)
```

```python
pd.read_sql('''select name from users where population = 115930''', conn)
```
